```python
ct = 0
for ind in df.index:
    path = df['path'][ind]
    name = df['name'][ind]
    x1 = df['xmin'][ind]
    x2 = df['xmax'][ind]
    y1 = df['ymin'][ind]
    y2 = df['ymax'][ind]
    try:
        img = cv2.imread(path)
        crop = img[y1:y2,x1:x2]
        os.makedirs("DogsImages/"+name, exist_ok=True)
        fl = "DogsImages/" + name + "/sample_" + str(ind+1) + ".jpg"
        cv2.imwrite(fl, crop)
    except:
        ct += 1
        # print(path)
        print(ct,name,path)
```
