```python
import os
import pandas as pd
```

```python
fold = os.listdir("DogsImages")
```

```python
head = ["path","dclass"]
field = []
for root, dirs, files in os.walk("DogsImages"):
	for file in files:
		fl_path = os.path.join(root, file)
		cls = fl_path.split("\\")[1]
		field.append([fl_path,cls])
field
```

```python
df = pd.DataFrame(data=field,columns=head)
df.head()
```

```python
df['category'] = df.dclass.astype('category').cat.codes
df.head(5)
```

```python
df['category'].value_counts()
```

```python
df.to_csv("stanford_dogsImages.csv", index=False)
```
